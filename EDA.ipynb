```python
x=df.drop('Species',axis=1)
```

```python
x
```

```python
y=df['Species']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
threshold=0.5
lr.fit(x_train,y_train)

pred1=lr.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train) 

pred2 = knn.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred2)
```

```python
from sklearn.svm import SVC
svn = SVC()
svn.fit(x_train, y_train)

pred3 = svn.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred3)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)

pred4 = dt.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred4)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rc = RandomForestClassifier()
rc.fit(x_train, y_train)

pred5 = rc.predict(x_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred5)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of Logistic Regression",accuracy_score(y_test,pred1))
print("Accuracy of Kneighbors",accuracy_score(y_test,pred2))
print("Accuracy of SVC",accuracy_score(y_test,pred3))
print("Accuracy of DecisionTree",accuracy_score(y_test,pred4))
print("Accuracy of RandomForest",accuracy_score(y_test,pred5))
```
